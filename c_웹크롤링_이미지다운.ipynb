```
import urllib.request as req
from bs4 import BeautifulSoup

import os #폴더생성
import urllib.parse as par #한글을 변환해줌
```

```
if not os.path.exists("./yahooimg"):
  os.mkdir("./yahooimg")

code = req.urlopen("https://images.search.yahoo.com/search/images;_ylt=Awr9Il4OSSFgT8AAA7FXNyoA;_ylu=Y29sbwNncTEEcG9zAzEEdnRpZAMEc2VjA3BpdnM-?p=%EB%91%98%EB%A6%AC&fr2=piv-web&fr=yfp-t")
soup = BeautifulSoup(code, "html.parser")
img = soup.select("li > a > img")
cnt = 1
for i in img:
  img_url = i.attrs["data-src"]
  req.urlretrieve(img_url, "./yahooimg/{}.png".format(cnt)) #이미지 다운로드
  print("이미지 크롤링 완료 {}".format(cnt))
  cnt += 1
```

```
keyword = input("키워드 입력>>")
encoded = par.quote(keyword)

if not os.path.exists("./yahooimg"):
  os.mkdir("./yahooimg")

if not os.path.exists("./yahooimg/{}".format(keyword)):
  os.mkdir("./yahooimg/{}".format(keyword))

url = "https://images.search.yahoo.com/search/images;_ylt=Awr9Il4OSSFgT8AAA7FXNyoA;_ylu=Y29sbwNncTEEcG9zAzEEdnRpZAMEc2VjA3BpdnM-?p={}&fr=yfp-t".format(encoded)
code = req.urlopen(url)
soup = BeautifulSoup(code, "html.parser")
img = soup.select("li > a > img")
for i in img:
  img_url = i.attrs["data-src"]
  try:
    req.urlretrieve(img_url, "./yahooimg/{}/{}.png".format(keyword, img.index(i)+1)) #이미지 다운로드
  except:
    print("이미지 없음")
  print("{} 이미지 크롤링 완료 {}".format(keyword, img.index(i)+1))
```
