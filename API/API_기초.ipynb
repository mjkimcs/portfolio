```python
from requests import Request, Session


clientId =  "QTQs6VtMu1RbNmWwMAbz"
clientSecret = "fLR602CF0m"
apiURL = "https://openapi.naver.com/v1/papago/n2mt"

s = Session()
headers = {
    "X-Naver-Client-Id" : clientId,
    "X-Naver-Client-Secret" : clientSecret
}

f = open("mydata.txt", encoding="utf-8")
f_out = open("mydata_res.txt", "w")
while True:
  line = f.readline()
  data = {
      "source":"ko",
      "target":"en",
      "text":line,
  }
  if not line:
    break
  print(line, end="\r")
  f_out.write(line)

  req = Request('POST', apiURL, data=data, headers=headers)
  draft = req.prepare()
  res = s.send(draft)
  res_json = res.json()
  print(res_json)
  try:
    translate = res_json["message"]["result"]["translatedText"]
    print(translate, end="\n\n")
    f_out.write(translate + "\n\n")
  except:
    pass

f_out.close()
f.close()
```
