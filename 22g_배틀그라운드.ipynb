## Feature Engineering

```
# 한 매치에서 100명 이하의 팀원들이 참가
# 100명이 참여한 매치에서 50등 한 것과 50명이 참여한 매치에서 50등 한 것은 실력차이가 있음
# 따라서, 매치에 참여한 참가자 수는 중요한 feature

df['players_num'] = df.groupby("matchId")["matchId"].transform("count")
df
```

```
# players_num을 고려한 정규화 (예를들어, 참가자가 적을수록 킬수를 높여주는 작업)
# 정규화 과정을 거친 값이 0이 나오는 것을 방지하기 위해 +1
# kills, damageDealt, maxPlace, matchDuration

df['killsNorm'] = df['kills']*((100-df['players_num'])/100+1)
df['damageDealtNorm'] = df['damageDealt']*((100-df['players_num'])/100+1)
df['maxPlaceNorm'] = df['maxPlace']*((100-df['players_num'])/100+1)
df['matchDurationNorm'] = df['matchDuration']*((100-df['players_num'])/100+1)

f = ['Id', 'kills', 'killsNorm', 'damageDealt', 'damageDealtNorm', 'maxPlace', 'maxPlaceNorm', 'matchDuration', 'matchDurationNorm']
df[f].head()
```

```
# 구급상자와 에너지드링크는 모두 작은 값이고
# 서로 비슷한 성질을 갖는 아이템을 사용한 횟수이기 때문에
# 합쳐서 좀 더 뚜렷한 값으로 만들어 줌

df['HealsBoosts'] = df['heals'] + df['boosts']
df[['heals', 'boosts', 'HealsBoosts']]
```

```
# 이상치 제거
# (1) 이동거리가 0인데, 킬이 1이상인 데이터 제거

df['totalDistance'] = df['rideDistance'] + df['walkDistance'] + df['swimDistance']
df['killsWithoutMoving'] = (df['totalDistance'] == 0) & (df['kills'] > 0) # T/F로 반환
df[df['killsWithoutMoving']==True]
```

## Random Forest

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```
X = df.drop(['winPlacePerc'], axis=1)
y = df['winPlacePerc']
```

```
X
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```
rf = RandomForestRegressor(n_estimators=40, min_samples_leaf=3, max_features='sqrt', n_jobs=-1)
rf.fit(X_train, y_train)
```

```
result = rf.predict(X_test)
error = mean_absolute_error(result, y_test)
error 
```

### 중요도가 높은 20개 컬럼으로만 학습시키기

```
fig = plt.figure(figsize=(15,5))

model_imp = pd.DataFrame(data=rf.feature_importances_, index=X.columns, columns=['Imp'])
model_imp.sort_values(by='Imp', ascending=False, inplace=True)
plt.bar(model_imp.index[:20],model_imp['Imp'][:20])

plt.xticks(rotation=90)
plt.show()
```

```
model_imp.index[:20]
```

```
X_f = X[model_imp.index[:20]].copy
```

```
type(model_imp.index[:20])
```

```
X_f()
```

```
X_train, X_test, y_train, y_test = train_test_split(X_f(), y, test_size=0.3, random_state=1)
```

```
rf = RandomForestRegressor(n_estimators=80, min_samples_leaf=3, max_features='sqrt', n_jobs=-1)
rf.fit(X_train, y_train)

result = rf.predict(X_test)
error = mean_absolute_error(result, y_test)
error 
```

```
rf = RandomForestRegressor(n_estimators=100, min_samples_leaf=3, max_features='sqrt', n_jobs=-1)
rf.fit(X_train, y_train)

result = rf.predict(X_test)
error = mean_absolute_error(result, y_test)
error 
```
