```python
package com.mytest;

public class Calc {

	private int a;
	private int b;
	private int hap;
	private int sub;
	private int mul;
	private int div;

	public Calc() {
		super();
		// TODO Auto-generated constructor stub
	}

	public Calc(int a, int b) {
		super();
		this.a = a;
		this.b = b;
	}

	public int getA() {
		return a;
	}

	public void setA(int a) {
		this.a = a;
	}

	public int getB() {
		return b;
	}

	public void setB(int b) {
		this.b = b;
	}

	public int getHap() {
		hap = a + b;
		return hap;
	}

	public int getSub() {
		sub = this.a - this.b;
		return sub;
	}

	public int getMul() {
		mul = this.getA() * this.getB();
		return mul;
	}

	public int getDiv() {
		try {
			div = this.getB() / getA();
		} catch (ArithmeticException AE) {
			this.a = 5;
			this.b = 10;
			System.out.println(AE.getMessage());
		} catch(Exception E) {
			System.out.println(E);
		}finally {
			System.out.println("반드시 처리해야 할 구문");
		}
		div = this.getB() / getA();
		return div;
	}

	@Override
	public String toString() {
		return String.format("Calc [a=%s, b=%s, getHap()=%s, getSub()=%s, getMul()=%s, getDiv()=%s]", a, b, getHap(),
				getSub(), getMul(), getDiv());
	}

	public static void main(String[] args) {
		Calc c1 = new Calc(); // 혹은 new Calc(20,100)
		System.out.println(c1);
	}

}
```

## 상속
1. 클래스 간의 상속 : 단일상속을 원칙
2. Interface(메뉴판/가공되지 않은 선언만 한 경우) 간의 상속 : 다중상속을 구현, 완벽한 추상화

- (아래)선조가 클래스(네모)이고, 후손이 Interface(동그라미)인 경우는 없음!

```python
# 선조
package com.test2;

public class AA {
	private int a;
	private int b;

	public AA() {
		System.out.println("AA의 생성자");
	}

	public int getA() {
		return a;
	}

	public void setA(int a) {
		this.a = a;
	}

	public int getB() {
		return b;
	}

	public void setB(int b) {
		this.b = b;
	}
	
	public int getHap() {
		return this.a + this.b;
	}

}
```

```python
# 후손
package com.test2;

public class BB extends AA {

	private int d;

	public BB() {
		System.out.println("BB");
	}

	public int getD() {
		return d;
	}

	public void setD(int d) {
		this.d = d;
	}

	public int getResult() {
		return getHap() - getD();
		// return (getA() + getB()) - getD();
		// return getHap() - d;
		// return getA() + getB() - this.d;
		// return super.getA() + super.getB() - this.d;
		// 안 됨! return a + b - d;
	}

}
```

```python
# 실행
package com.test2;

public class Test01 {
	
	public static void main(String[] args) {
		BB b1 = new BB();
	}
}


# 결과
AA의 생성자
BB
```

```python
# 선조
package com.test2;

public class AA {
	private int a;
	private int b;

	public AA() {
		System.out.println("AA의 생성자");
	}

	protected int getA() {
		return a;
	} 
	// protected 의미
	// 1. 나를 상속받는 애들은 public처럼 다 가져다가 써라~
	// 2. 클래스 변수(b1)는 같은 패키지에서만 public처럼 호출이 가능

	public void setA(int a) {
		this.a = a;
	}

	protected int getB() {
		return b;
	}

	public void setB(int b) {
		this.b = b;
	}
	
	public int getHap() {
		return this.a + this.b;
	}

}

```

```python
# 후손
package com.test2;

public class BB extends AA {

	private int d;

	public BB() {
		System.out.println("BB");
	}

	public int getD() {
		return d;
	}

	public void setD(int d) {
		this.d = d;
	}

	public int getResult() {
		// return getHap() - getD();
		// return (getA() + getB()) - getD();
		// return getHap() - d;
		return getA() + getB() - this.d;
		// return super.getA() + super.getB() - this.d;
		// 안 됨! return a + b - d;
	}

}
  
```

```python
# 실행
# protected 때문에 다른 패키지에서는 실행이 안 됨!
package com.test2;

public class Test01 {
	
	public static void main(String[] args) {
		BB b1 = new BB();
		
		b1.setA(100);
		b1.setB(200);
		
		System.out.println(b1.getHap());
		System.out.println(b1.getResult());
		System.out.println(b1.getA() + ":" + b1.getB());
	}
}


# 결과
AA의 생성자
BB
300
300
100:200
```
