## DOM(Document Object Model)
- html, xml 문서에 대한 프로그래밍 인터페이스
- 객체지향 모델로서 구조화된 문서를 표현하는 형식
- document : DOM의 최고 상위 객체
- element : html 문서에 사용되는 모든 요소와 태그, attribute를 포함
- attribute : element가 가진 속성
- node : 객체를 지칭, 요소/속성/텍스트 등을 말함  
```
  - nodeName : 노드의 이름을 리턴
  - nodeType : 노드의 유형을 정수로 리턴(1:요소, 2:속성, 3:텍스트, 8:주석)
  - childNodes : 노드의 자식노드 리스트
  - firstChild : 노드의 첫번째 자식노드
  - lastChild : 노드의 마지막 자식노드
  - previousSibling : 노드의 이전 위치에 자리한 형제노드
  - nextSibling : 노드의 다음 위치에 자리한 형제노드
  - parentNode : 노드의 부모노드
```
