```python
# 수치형변수 중 `time`은 정규분포에 가깝고, 나머지는 멱함수(power-law)분포에 가까움
sns.pairplot(data=df, vars=num_cols + [target_col], size=4)
```

## 선형회귀 모델 학습

### RMSE/RMSLE 손실함수 정의

```python
rmse = lambda y, p: np.sqrt(mean_squared_error(y, p))
rmsle = lambda y, p: np.sqrt(mean_squared_error(np.log1p(y), np.log1p(p)))
```

### 변수 가공없이 선형회귀 모델 학습

```python
lr = LinearRegression()
lr.fit(df[num_cols], df[target_col])
df[pred_col] = lr.predict(df[num_cols])

print(f' RMSE:\t{rmse(df[target_col], df[pred_col]):.2f}')
```

```python
# 종속변수가 멱함수 분포라서 예측값에 음수가 존재하고 큰 값의 예측은 비교적 정확한 반면 적은 값의 예측은 오차가 많이 발생함
# 따라서 log1p 변환을 해줘야 함
sns.pairplot(data=df, vars=[pred_col, target_col], size=4, plot_kws={'alpha': .5})
```

### 멱함수 분포 변수 `log1p` 변환

```python
# np.log1p() 함수로 정규분포에 가깝게 변환
# 역변환은 np.expm1() 함수를 이용
df["box_off_num"].apply(np.log1p).hist(bins=50)
```

```python
lr = LinearRegression()
lr.fit(df[num_cols], np.log1p(df[target_col])) ##
df[pred_col] = np.expm1(lr.predict(df[num_cols])) ##

print(f' RMSE:\t{rmse(df[target_col], df[pred_col]):.2f}')
print(f'RMSLE:\t{rmsle(df[target_col], df[pred_col]):.2f}')
```

```python
sns.pairplot(data=df, vars=[pred_col, target_col], size=4, plot_kws={'alpha': .5})
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df[target_col], df[pred_col], alpha=.5)
plt.xlabel('box_off_num')
plt.ylabel('prediction')
plt.yscale('log') ##
plt.xscale('log') ##
```

```python
df[num_cols].describe()
```

```python
# 수치형 독립변수 중 멱변환 분포를 따르는 변수에도 `np.log1p()` 변환을 적용
df[['dir_prev_bfnum', 'dir_prev_num', 'num_staff', 'num_actor']] = df[['dir_prev_bfnum', 'dir_prev_num', 'num_staff', 'num_actor']].apply(np.log1p)
df[num_cols].describe()
```

```python
sns.pairplot(data=df, vars=num_cols, size=5)
```

```python
lr = LinearRegression()
lr.fit(df[num_cols], np.log1p(df[target_col]))
df[pred_col] = np.expm1(lr.predict(df[num_cols]))

# RMSLE는 조금 나빠졌지만 RMSE는 크게 개선됨
print(f' RMSE:\t{rmse(df[target_col], df[pred_col]):12.2f}')
print(f'RMSLE:\t{rmsle(df[target_col], df[pred_col]):12.2f}')
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df[target_col], df[pred_col], alpha=.5)
plt.xlabel('box_off_num')
plt.ylabel('prediction')
plt.yscale('log')
plt.xscale('log')
```

### 날짜/시간 변수 처리

```python
df['release_time'] = pd.to_datetime(df['release_time'])
df['release_time'].describe()
```

```python
df['year'] = df['release_time'].dt.year
df['month'] = df['release_time'].dt.month
df.head()
```

```python
num_cols += ['year', 'month']
print(num_cols)
```

```python
lr = LinearRegression()
lr.fit(df[num_cols], np.log1p(df[target_col]))
df[pred_col] = np.expm1(lr.predict(df[num_cols]))

print(f' RMSE:\t{rmse(df[target_col], df[pred_col]):.2f}')
print(f'RMSLE:\t{rmsle(df[target_col], df[pred_col]):.2f}')
```

### 독립변수 정규화/스케일링
- 딥러닝과 비선형 알고리즘에서는 정규화가 반드시 필요하지만,
- 선형회귀나 결정트리 류의 알고리즘에서는 도움이 안 됨
- 아래에서 알 수 있듯이 스케일링을 해주기 전과 후의 오차값이 동일함

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# 변수에서 평균을 빼준 후 표준편차로 나눠줌
scaler = StandardScaler()
X = scaler.fit_transform(df[num_cols])

lr = LinearRegression()
lr.fit(X, np.log1p(df[target_col]))
df[pred_col] = np.expm1(lr.predict(X))

print(f' RMSE:\t{rmse(df[target_col], df[pred_col]):12.2f}')
print(f'RMSLE:\t{rmsle(df[target_col], df[pred_col]):12.2f}')
```

```python
# 0~1로 변환
scaler = MinMaxScaler()
X = scaler.fit_transform(df[num_cols])

lr = LinearRegression()
lr.fit(X, np.log1p(df[target_col]))
df[pred_col] = np.expm1(lr.predict(X))

print(f' RMSE:\t{rmse(df[target_col], df[pred_col]):12.2f}')
print(f'RMSLE:\t{rmsle(df[target_col], df[pred_col]):12.2f}')
```

### 2차 다항회귀 (Polynomial Regression)
- 오버피팅의 위험이 있으므로 3차 이상의 다항회귀는 추천하지 않음

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(2)
X = poly.fit_transform(df[num_cols])
print(X.shape)
```

```python
lr = LinearRegression()
lr.fit(X, np.log1p(df[target_col]))
df[pred_col] = np.expm1(lr.predict(X))

print(f' RMSE:\t{rmse(df[target_col], df[pred_col]):12.2f}')
print(f'RMSLE:\t{rmsle(df[target_col], df[pred_col]):12.2f}')
```
