# 선형회귀

- 각각의 독립변수와 종속변수가 서로 선형성만 만족하면 그것은 선형회귀(3x_3 + 4x_2 + 2x + 5)
- 다항회귀(Polynomial, 3x^3 + 4x^2 + 2x + 5) 식을 살짝 바꿔보면 다중회귀(Multi Linear, 3x_3 + 4x_2 + 2x_1 + 5x_0)
- 집합으로 표현하면, 다항회귀 > 다중회귀 > 단순선형회귀
- 즉, 곡선으로 그려져도 선형성만 만족하면 그것은 선형회귀
- Logistic Regression은 비선형회귀

## 학습데이터 로드

```python
data_dir = Path('./input')
trn_file = data_dir / 'train.csv'
seed = 42
```

```python
trn = pd.read_csv(trn_file, index_col=0)
print(trn.shape)
trn.head()
```

```python
sns.pairplot(trn.sample(n=10000), vars=['u', 'dered_u'], hue='class', size=5)
```

```python
X = trn[trn['class'] != 0][['u']].values
y = trn[trn['class'] != 0]['dered_u'].values
print(X.shape, y.shape)
```

## RMSE 손실함수 정의

```python
# root를 씌워주는 이유는 scale을 맞춰줘서 해석하기 용이함

rmse = lambda y, p: np.sqrt(np.mean((y - p) ** 2))

# def rmse(y, p):
#     return np.sqrt(np.mean((y-p) ** 2))
```

## NumPy를 이용한 선형회귀

```python
alpha = .001
theta0 = .1
theta1 = .1
epoch = 4
```

```python
x = X.flatten() # 2차원 -> 1차원
print(x.shape)
```

```python
rmses = []
for i in tqdm(range(epoch)):
    for x_i, y_i in zip(x, y):
        p_i =  theta0 + theta1 * x_i
        theta0 = theta0 + alpha * (y_i - p_i)
        theta1 = theta1 + alpha * (y_i - p_i) * x_i
    rmses.append(rmse(y, theta0 + theta1 * x))
```

```python
theta1, theta0
```

```python
rmses
```

```python
plt.plot(rmses)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X, y, alpha=.1)
plt.plot(X, theta0 + theta1 * x, color='darkorange', linewidth=.5)
plt.xlim(0, 40)
plt.ylim(0, 40)
plt.xlabel('u')
plt.ylabel('dered_u')
```

## scikit-learn을 이용한 선형회귀

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
p = lr.predict(X)
```

```python
rmse(y, p)
```

```python
lr.coef_, lr.intercept_
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X, y, alpha=.1)
plt.plot(X, p, color='darkorange', linewidth=.5)
plt.xlim(0, 40)
plt.ylim(0, 40)
plt.xlabel('u')
plt.ylabel('dered_u')
```

## scikit-learn을 이용한 로지스틱회귀
- 손실함수가 Linear Regression과 다르지만,
- Gradient Descent(경사하강법)로 세타를 업데이트하는 수식이 Linear Regression과 동일함

```python
X = trn[trn['class'] != 0].drop('class', axis=1)
y = trn[trn['class'] != 0]['class'].values - 1
```

```python
np.unique(y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score

lr = LogisticRegression()
lr.fit(X, y)
p = lr.predict_proba(X)[:, 1]
```

```python
p_test = lr.predict_proba(X)
p_test
```

```python
lr.coef_, lr.intercept_
```

```python
log_loss(y, p)
```

```python
roc_auc_score(y, p)
```

```python
from sklearn.metrics import plot_roc_curve
from sklearn.metrics import plot_precision_recall_curve
```

```python
plt.figure(figsize=(8, 8))
plot_roc_curve(lr, X, y);
plt.title('ROC Curve');
```

```python
plt.figure(figsize=(8, 8))
plot_precision_recall_curve(lr, X, y);
plt.title('Precision-Recall Curve');
```
