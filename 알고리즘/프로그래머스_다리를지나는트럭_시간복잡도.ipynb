- O(logN) : N이 16일 때 4
- O(N)
- O(NlogN) : N이 16일 때 16*4 = 64
- O(N^2) : 16^2 = 256

```python
def solution(bridge_length, weight, truck_weights):
  time = 0
  ing = [0] * bridge_length  # bridge_length가 2라면 [0, 0]

  while ing:
    time += 1
    ing.pop(0)
    if truck_weights:  # truck_weights가 []라면 break
      if truck_weights[0] + sum(ing) <= weight:
        ing.append(truck_weights.pop(0))
      else:
        ing.append(0)    
  return time
```

```python
solution(2, 10, [7,4,5,6])
```

```python
solution(100, 100, [10])
```

```python
# deque 써보기

from collections import deque

def solution(bridge_length, weight, truck_weights):
  time = 0
  ing = [0] * bridge_length
  ing = deque(ing)
  truck_weights = deque(truck_weights)

  while ing:
    time += 1
    ing.popleft()
    if truck_weights:
      if truck_weights[0] + sum(ing) <= weight:
        ing.append(truck_weights.popleft())
      else:
        ing.append(0)    
  return time
```

```python
solution(100, 100, [10,10,10,10,10,10,10,10,10,10])
```
