```python
def solution(N, stages):
  result = {}
  length = len(stages)
  for i in range(1, N+1):
    if length != 0:
      count = stages.count(i)
      result[i] = count/length
      length -= count
    else:
      result[i] = 0
  return sorted(result, key=lambda x:result[x], reverse=True)
```

```python
solution(5, [2,1,2,6,2,4,3,3])
```

```python
solution(4, [4,4,4,4,4])
```
