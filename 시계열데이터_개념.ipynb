# 1. 시계열 데이터(Time Series Data)란?
*  개념: 데이터를 시간의 흐름에 따라 나열한 것
*  특성: 지속적으로 업데이트되는 데이터, 동일한 시간 간격으로 관찰된 데이터
*  활용: 미래예측, 2개 이상의 시계열 간의 상관관계를 통해 서로 어떤 영향을 주는지 분석

# 2. 시계열 데이터의 분류
*  Trend: 전반적으로 증가 또는 감소하는 패턴
*  Seasonality: 규칙적인 주기에 따라 반복되는 패턴
*  Cyclicity: 주기가 규칙적이지 않지만 반복되는 패턴
*  Random noise: 불규칙한 파동
