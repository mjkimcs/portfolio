### 학습데이터와 테스트데이터 분리


```python
from math import floor
```

```python
def train_test_split_df(df, test_size):
  div = floor(df.shape[0] * (1 - test_size))
  return df.loc[:div], df.loc[div + 1:]
```

```python
train_df, test_df = train_test_split_df(df, 0.1)
```

```python
train_df.tail()
```

## Prophet 모델 학습

```python
from fbprophet import Prophet
```

```python
model = Prophet(changepoint_range=0, changepoint_prior_scale=0.0001, seasonality_mode='multiplicative', holidays_prior_scale=0.05, yearly_seasonality=True)
model.fit(train_df)
```

### 학습 결과 시각화

```python
from fbprophet.plot import add_changepoints_to_plot
```

```python
pred = model.predict(test_df)
```

```python
# pred = model.predict(test_df)
f = model.plot(pred)
plt.plot(test_df['ds'], test_df['y'], 'g-', label='actual')
add_changepoints_to_plot(f.gca(), model, pred)
plt.legend(loc='lower right')
plt.show()
```

```python
# 정현님 시작
def predict_and_test(range, scale, season) :
  #model = Prophet(changepoint_range=range, changepoint_prior_scale=scale, seasonality_mode='multiplicative', holidays_prior_scale=0.05, yearly_seasonality=True)
  model = Prophet(changepoint_range=range, changepoint_prior_scale=scale, seasonality_mode='additive', seasonality_prior_scale=season)
  model.fit(train_df)
  pred = model.predict(test_df)
  result = r2_score(test_df['y'], pred['yhat'])
  print('R2 Score:', range,  scale, season, result)
  return result
```

```python
r2_result = pd.DataFrame(columns=['range' , 'scale', 'season', 'result'])

for ra in range(8 ,11):  # 0.05, 0.1, 0.15 ~10.0
  for sc in range(1 ,21):  # 0.05, 0.1, 0.15 ~10.0
    for se in range(0, 3): #0,10,20,30
      p_result = predict_and_test(ra *0.1 , sc*0.05, se*10)
      r2_result=r2_result.append({'range' : ra * 0.1 , 'scale' : sc * 0.05 , 'season' : se * 10, 'result': p_result} , ignore_index=True)

```

```python
r2_result.sort_values(by=['result'], axis=0, ascending=False)
```

# SCFI

```python
df = pd.read_csv('/content/SCFI_ (1).CSV', encoding='CP949')
df['ds'] = df['ds'].apply(lambda s: pd.to_datetime(str(s)))
df.head(3)
```

```python
from math import floor

def train_test_split_df(df, test_size):
  div = floor(df.shape[0] * (1 - test_size))
  return df.loc[:div], df.loc[div + 1:]

train_df, test_df = train_test_split_df(df, 0.1)
```

```python
from fbprophet import Prophet
from fbprophet.plot import add_changepoints_to_plot
from sklearn.metrics import r2_score

model = Prophet(changepoint_range=0.8, changepoint_prior_scale=0.05, seasonality_mode='additive')
model.fit(train_df)
pred = model.predict(test_df)
```

```python
def predict_and_test(range, scale, season) :
  #model = Prophet(changepoint_range=range, changepoint_prior_scale=scale, seasonality_mode='multiplicative', holidays_prior_scale=0.05, yearly_seasonality=True)
  model = Prophet(changepoint_range=range, changepoint_prior_scale=scale, seasonality_mode='additive', seasonality_prior_scale=season)
  model.fit(train_df)
  pred = model.predict(test_df)
  result = r2_score(test_df['y'], pred['yhat'])
  print('R2 Score:', range,  scale, season, result)
  return result

r2_result = pd.DataFrame(columns=['range' , 'scale', 'season', 'result'])

for ra in [0.8, 0.95]:
  for sc in [0.001, 0.005, 0.01, 0.05, 0.1, 0.5]:
    for se in [0.01, 0.05, 0.1, 0.5, 1, 2, 3, 5, 7, 9 ,10]:
      p_result = predict_and_test(ra, sc, se)
      r2_result=r2_result.append({'range' : ra , 'scale' : sc, 'season' : se, 'result': p_result} , ignore_index=True)
```

```python
r2_result.sort_values(by=['result'], axis=0, ascending=False)
```

```python
range, scale, season = 0.8, 0.05, 10
predict_and_test(range, scale, season)

model = Prophet(changepoint_range=range, changepoint_prior_scale=scale, seasonality_mode='additive', seasonality_prior_scale=season, yearly_seasonality=True)
model.fit(train_df)
pred = model.predict(test_df)
result = r2_score(test_df['y'], pred['yhat'])
print(result)
```

```python
from fbprophet.plot import add_changepoints_to_plot

pred = model.predict(test_df)
```

```python
from sklearn.metrics import r2_score

print('R2 Score: ', r2_score(test_df['y'], pred['yhat']))
```

```python
f = model.plot(pred)
plt.plot(test_df['ds'], test_df['y'], 'g-', label='actual')
add_changepoints_to_plot(f.gca(), model, pred)
plt.legend(loc='lower right')
plt.show()
```

### 학습 결과 평가 : R2 Score
- 1에 가까울수록 예측력 높음  
- 음수로 갈수록 예측력 낮음

```python
from sklearn.metrics import r2_score

print('R2 Score: ', r2_score(test_df['y'], pred['yhat']))
```
