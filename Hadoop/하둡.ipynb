## 형변환
    byte(1) -> short(2) or char(2) -> int(4) -> long(8) -> float(4) -> double(8)

1. 명시적 형변환 : int a = (int)90.8; //90이 되어서 int방에 들어감
2. 묵시적 형변환 : double d = 90; //d=90.0

```python
package com.test;

public class Test03 {
	
	public static void main(String[] args) {
        int month = 8; 
        String monthString;
        switch (month) {  //month 자리에 4바이트 이하인 것들은 형변환 가능
            case 1:  monthString = "January";
                     break;
            case 2:  monthString = "February";
                     break;
            case 3:  monthString = "March";
                     break;
            case 4:  monthString = "April";
                     break;
            case 5:  monthString = "May";
                     break;
            case 6:  monthString = "June";
                     break;
            case 7:  monthString = "July";
                     break;
            case 8:  monthString = "August";
                     break;
            case 9:  monthString = "September";
                     break;
            case 10: monthString = "October";
                     break;
            case 11: monthString = "November";
                     break;
            case 12: monthString = "December";
                     break;
            default: monthString = "Invalid month";
                     break;
        }
        System.out.println(monthString);
	}


}
```

```python
# for문

package com.test;

public class Test01 {
	
	public static void for_test() {
		System.out.println("casee 1 : 1~10 출력");
		
		int i = 0;
		for(i=1; i<=10; i++) {
			System.out.printf("%5d", i);
		}
		System.out.println("\n i=" + i);
	}
	
	public static void for_test02() {
		System.out.println("\n case 2 :10~1 출력");
		
		for(int i=10; i>0; i--) {
			System.out.printf("%5d", i);
		}
	}

	
	public static void main(String[] args) {
		for_test();
		
		int a = 10;
		int k = 300;
		{
			System.out.println("a=" + a);
			k = 100; //재대입
			System.out.println("k=" + k);
		}
		System.out.println("k=" + k);
	}
}

```

```python
# 중첩 for
package com.test;

public class Test01 {

	public static void for_test04() {
		System.out.println("\n case 4 : 중첩 for 출력");
		
		for(int i=1; i<=25; i+=5) {
			for(int j=i; j<=i+4; j++) {
				System.out.printf("%5d", j);
			}
			System.out.println();
	}
}

	public static void main(String[] args) {
		for_test04();
	}

}

```

```python
# 중첩 for
package com.test;

public class Test01 {

	public static void for_test04() {
		System.out.println("\n case 4 : 중첩 for 출력");
		
		for(int i=25; i>=21; i--) {
			for(int j=i; j>=i-20; j-=5) {
				System.out.printf("%5d", j);
			}
			System.out.println();
	}
}

	public static void main(String[] args) {
		for_test04();
	}

}

# 결과
 case 4 : 중첩 for 출력
   25   20   15   10    5
   24   19   14    9    4
   23   18   13    8    3
   22   17   12    7    2
   21   16   11    6    1
```
