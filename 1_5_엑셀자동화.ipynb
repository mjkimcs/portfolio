```
import openpyxl
from openpyxl.styles import Font, Alignment, PatternFill, Border, Color, Side
from openpyxl.chart import Reference, Series, LineChart, BarChart, PieChart
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```
df = pd.read_excel("./아파트분양가격.xlsx")
df["지역명"].unique()
```

```
print(df["지역명"].unique())
```

```
locations = df["지역명"].unique()
for i in locations:
  
  writer = pd.ExcelWriter("./아파트분양가격_{}.xlsx".format(i)) #pd.ExcelWriter: 엑셀을 좀 더 풍성하게 사용가능
  
  df_loc = df[df["지역명"] == i]
  years = df_loc["연도"].unique()
  for y in years:
    df_year = df_loc[df_loc["연도"] == y]
    df_result = df_year.sort_values(by="월", ascending=True)
    df_result.to_excel(writer, sheet_name="{}년".format(y), index=None) #writer파일에 분리된 데이터를 각 시트별로 기입
  writer.save()
  print("{}지역 분리 완료".format(i))

#####서식
user_font = Font(name="맑은 고딕", size=12, bold=True)
user_alignment = Alignment(horizontal="center") #가운데정렬
orange_color = PatternFill(patternType="solid", fgColor=Color("FE9A2E")) #색깔코드 구글링
gray_color = PatternFill(patternType="solid", fgColor=Color("BDBDBD"))
user_border = Border(left=Side(style="thin"), right=Side(style="thin"), top=Side(style="thin"), bottom=Side(style="thin"))
#########

for i in locations:
  book = openpyxl.load_workbook("./아파트분양가격_{}.xlsx".format(i))
  for s in book.sheetnames: #book.sheetnames는 sheet이름을 리스트 자료형으로 반환
    sheet = book[s]
    for row in sheet["A1:E1"]: #헤더 서식지정
      for cell in row:
        cell.font = user_font
        cell.alignment = user_alignment
        cell.fill = orange_color
        cell.border = user_border
    
    for row in sheet["A2:E{}".format(sheet.max_row)]: #본문 서식지정
      for cell in row:
        cell.alignment = user_alignment
        cell.fill = gray_color
        cell.border = user_border

    chart = BarChart() #차트 그리기
    chart.title = "{}지역 {}년도 아파트 분양가격".format(i, s)
    value = Reference(sheet, range_string="{}!E2:E{}".format(s, sheet.max_row))
    value_series = Series(value, title="분양가격")
    chart.append(value_series)
    sheet.add_chart(chart, "G1")

  book.save("./아파트분양가격_{}.xlsx".format(i))
  print("{}지역 서식지정 완료".format(i))
```

```
import pandas as pd

writer = pd.ExcelWriter("./테스트_합치기.xlsx") #엑셀을 좀 더 풍성하게 사용가능
for year in range(2017, 2021):
  df = pd.read.excel("./테스트{}.xlsx".format(year), sheet_name="sheet1") #특정 sheet만 불러오기
  df[["시설명", "주소"]] #2개 열만 불러오기
  df[["시설명", "주소"]].to_excel(writer, sheet_name="{}년".format(year)) #sheet이름 수정하기
writer.save()
```

```
#엑셀서식

import openpyxl
from openpyxl.styles import Font, Alignment, PatternFill, Border, Color, Side

book = openpyxl.load_workbook("./2017년_광고비_total.xlsx")
sheet = book.active

user_font = Font(name="맑은 고딕", size=12, bold=True)
user_alignment = Alignment(horizontal="center") #가운데정렬
orange_color = PatternFill(patternType="solid", fgColor=Color("FE9A2E")) #색깔코드 구글링
gray_color = PatternFill(patternType="solid", fgColor=Color("BDBDBD"))
user_border = Border(left=Side(style="thin"), right=Side(style="thin"), top=Side(style="thin"), bottom=Side(style="thin"))

sheet["A14"].value = "합계"
sheet["A14"].font = user_font
sheet["A14"].alignment = user_alignment
sheet["A14"].fill = orange_color #셀 색
sheet["A14"].border = user_border 

sheet["B14"].value = "=SUM(B2:B13)"
sheet["C14"].value = "=SUM(C2:C13)"
sheet["D14"].value = "=SUM(D2:D13)"

for row in sheet["B2:D13"]: #하나의 행 가져오기
  for cell in row: #하나의 셀 가져오기
    cell.font = user_font
    cell.alignment = user_alignment
    cell.fill = gray_color
    cell.border = user_border

book.save("./2017년_광고비_total_서식지정.xlsx")
```

```
#차트그리기

import openpyxl 
from openpyxl.chart import Reference, Series, LineChart, BarChart, PieChart

company = ["LG전자", "삼성전자", "현대자동차"]
for i in company:
  book = openpyxl.load_workbook("./2017년_광고비_{}.xlsx".format(i))
  sheet = book.active #sheet = book["Sheet1"]
  chart = LineChart()
  chart.title = "{} 월별 광고비".format(i)

  for c in ["C", "D", "E", "F", "G"]:
    value = Reference(sheet, range_string="Sheet1!{}2:{}13".format(c, c))
    value_series = Series(value, title=sheet["{}1".format(c)].value)
    chart.append(value_series)
  
  sheet.add_chart(chart, "J1")
  book.save("./2017년_광고비_{}_차트.xlsx".format(i))
```
