-  dll(dynamic link library): 참조형 실행파일, 참조 당하는 파일이기 때문에 실행메모리를 차지하지 않음, 즉 exe가 실행할 때 dll을 참조, dll을 넣고 exe를 만들어야 메모리가 작아서 빠름, .py 1개이상 여러 개를 가지고 있음(.pyd는 .py 1개만)
-  exe: 그냥 실행파일

-  tcl : GUI
-  Tools : 툴 실행하는 명령들
