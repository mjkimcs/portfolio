## /robots.txt로 법적으로 페이지 접근 가능여부를 볼 수 있음

```python
from bs4 import BeautifulSoup
import urllib.request

def Test02():
  list_result01 = []
  list_result02 = []
  with open("mytest.txt", "w", encoding="utf-8") as f:
    for i in range(1, 2):
      url = "https://home.ebs.co.kr/ladybug/board/6/10059819/oneBoardList?c.page=" + str(i) + "&hmpMnuId=106&searchCondition=&searchConditionValue=0&searchKeywordValue=0&searchKeyword=&bbsId=10059819&"
      url = urllib.request.Request(url)
      result = urllib.request.urlopen(url).read().decode("utf-8")
      
      soup = BeautifulSoup(result, "html.parser")
      result01 = soup.find_all("span", class_="date")
      result02 = soup.find_all("p",class_="con")

      for i in result01:
        list_result01.append(i.get_text(" ", strip=True))
      for i in result02:
        list_result02.append(i.get_text(" ", strip=True))

    for k, h in zip(list_result01, list_result02):
      f.write(k + ' ' + h + '\n')
      print(k, h)

if __name__ == '__main__':
    Test02()
```

```python
from bs4 import BeautifulSoup
import urllib.request


def Test():
  url = 'https://www.donga.com/news/search?p=1&query=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&check_news=1&more=1&sorting=1&search_date=1&v1=&v2=&range=1'
  url = urllib.request.Request(url)
  result = urllib.request.urlopen(url).read().decode("utf-8")

  soup = BeautifulSoup(result, "html.parser")
  result01 = soup.find_all("p", class_="txt")
  links = []
  for i in result01:
    for k in i:
      links.append(k.get("href"))
  return links


def Test2():
  links = Test()
  params = []
  for i in links:
    url = urllib.request.Request(i)
    result = urllib.request.urlopen(url).read().decode("utf-8")

    soup = BeautifulSoup(result, "html.parser")
    result01 = soup.find_all("div", class_="txt_ban")
    params = []
    for i in result01:
      params.append(i.get_text(" ", strip=True))
  print(params)


if __name__ == '__main__':
    print(Test())
    Test2()
```

```python
from bs4 import BeautifulSoup
import urllib.request
from collections import Counter


def Test():
  with open("mytest.txt", encoding="utf-8") as f:
    res = f.read().split()
    res.sort()
    
    countlist = Counter(res)
    print(countlist)

    res2 = countlist.most_common(5)
    print(res2)
    for i in res2:
      print(i[0], i[1])


if __name__ == '__main__':
    Test()
```
