**클래스**
-	클래스 = 변수(Variable) + 메소드(Methods)
-	객체지향프로그램(OOP, Object Oriented Programming)의 사용자정의 데이터형
-	실제로 존재하는 사물의 속성(변수)과 기능(메소드)을 모델링해서 추상화시키는 과정
- 캡슐화: 오픈된 메소드가 은닉된 멤버변수에게 값 전달 및 변경하는 구조

**객체**
-	객체 = 속성(Attrivutes) + 행위(Behaviors)
- 속성을 변경하는 것이 행위
-	모듈
-	자신 고유의 속성(키와 값) + 클래스에서 정의한 행위를 수행(움직인다)
-	정의된 클래스를 사용해서 실제로 메모리에 생성되어 메모리에 로딩된 상태
-	객체는 클래스의 타입으로 선언됐을 때를 의미하는 것이고, 그 객체가 메모리에 할당되어 실제 사용될 때 인스턴스라고 함
- 객체를 생성했을 경우 비로소 인스턴스가 됨
-	객체는 클래스의 인스턴스

**상속**
* 후손(this) 클래스 -> 선조(super/구체/concrete/부모) 클래스
* 선조는 후손을 모름
* 후손은 선조를 알고 상속 받는다
* 다중상속: 후손은 선조 2개를 상속 받을 수 있음
* 후손은 바로 위 선조만 모시면 됨
* 후손은 상속을 받아서 기능을 확장
* 생성자, 소멸자, 특정메소드(_ _ init _ _ ), 강한 private(_ _ b)은 상속받을 수 없음
* 상속의 목적: 재사용, 간결성, 확장성

## 상속(inheritance)
-  생성자, 소멸자, 특정메소드(_ _ init _ _ ), 강한 private(_ _ b)은 상속받을 수 없음
-  상속의 목적은 재사용, 간결성, 확장성

class.png

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def personInfo(self):
    return self.name + "(age :" + str(self.age) + ")"


class Student(Person):
  def __init__(self, name, age, grade):
    Person.__init__(self, name, age)
    self.grade = grade # grade

  def studentInfo(self):
    return self.personInfo() + ",grade :" + str(self.grade)



if __name__ == '__main__':
  x = Student("민정", 20, 2)
  # 객체를 생성하면
  # 힙영역에 선조클래스(&100)가 생성된 후 후손클래스(&200)가 생성 그 다음 화살표로 연결(선조<-후손)
  # <<번외: 후손클래스가 주소(&200)만 받은 상태(완공되기 전)에서 선조를 참조 후 grade가 끝나고 후손클래스가 완전히 생성>>
  # stack&static 영역에 x가 생성되고 선조의 주소(&100)를 받아서 대입시킨 후 선조를 참조
  # 선조의 주소를 따라가면 studentInfo()를 호출할 수 있음

  print(x.studentInfo())
  print(x.personInfo())
```

## 재정의
-  선조의 메소드와 후손의 메소드가 같을 경우 선조의 메소드를 히든 처리
-  선조 클래스에 이미 정의되어 있는 메소드의 기능을 변경해서 새로 정의하는 것
-  선조가 가진 메소드의 파라미터 갯수가 같아야 함
-  리턴형은 같지 않아도 됨

```python
class MyScore:
  def __init__(self, kor, eng):
    self.kor = kor
    self.eng = eng
  def getTot(self):
    print("선조")
    return(self.kor + self.eng)

class MyScore_Sub(MyScore):
  def __init__(self, kor, eng, mat):
    super().__init__(kor, eng)
    self.mat = mat
  def getTot(self):
    print("후손")
    return(super().getTot() + self.mat)


if __name__ == '__main__':
  print(MyScore(20, 30).getTot())
  print("=======")
  print(MyScore_Sub(100, 100, 100).getTot()) 
  # 재정의된 후손의 메소드가 호출
  # 외부에서 호출될 때만 선조의 메소드를 히든 처리
  # 객체가 봤을 때 선조의 주소번지가 히든이라는 의미
  # 후손은 선조가 당연히 보임
```

## 연산자 오버로딩
- object가 주는 메소드
- 정렬 시 필요


```python
class MYSU:
  def __init__(self, Su):
    self.Su = Su
  def __lt__(self, other): # 오른쪽이 클 경우, self는 a / other는 b
    print(self.Su < other.Su)  # 100 < 50
  def __gt__(self, other):# 왼쪽이 클 경우
    print(self.Su > other.Su)
  def __eq__(self, other):# 같을 경우
    print(self.Su == other.Su)

if __name__ == '__main__':
  a = MYSU(100)
  b = MYSU(50)
  a < b
  a > b
  a == b
```

# 다형성
- Subclassing(그냥 상속)과 Subtyping(추상화)
- 다양한 형태의 성질로 이루어진 클래스의 동적바인딩 구조
- 하나의 선조(Base)에 여러 개의 후손(각기 다른 기능)이 있음
- 상속 시에 같은 이름의 메소드가 여러 클래스에서 다른 기능을 하는 것
- 후손 클래스에서 선조 클래스의 메소드를 **재정의**


## 1. Subclassing: 그냥 상속
* 후손 클래스는 무조건 재정의
* 상속의 유형 중에서 선조 클래스의 속성과 구조를 그대로 가지면서 새로운 멤버 변수와 멤버 메소드를 추가해서 재사용하는 방법

## 2. Subtyping: 추상화(abstract)
* 후손 클래스는 무조건 재정의
* 선조 클래스(추상 클래스)는 추상 메소드(메소드명이 있지만 기능은 없음/후손을 위해 선조가 강제로 재정의를 할 수 있게끔 선언된 메소드)를 포함
* 추상 클래스는 객체 생성이 불가능  
* 추상 클래스를 상속받는 후손 클래스가 선조의 추상 메소드를 재정의하지 않으면 추상 클래스가 됨
* duck typing(동적 바인딩의 한 종류)을 지원

duc.png

* @abstractmethod: 메소드를 추상메소드로 지정
* @property: 메소드를 변수로 호출할 수 있도록 지정
* @staticmethod: 정적메소드로 지정
* @classmethod: 클래스 인스턴스 메소드로 지정

```python
from abc import abstractmethod, ABCMeta

class Base(metaclass = ABCMeta):
  
  @abstractmethod
  def Start(self):
    print("base start")

  @abstractmethod
  def Stop(self):
    print("base stop")


class Cat(Base):
  def Start(self):
    print("cat start")
  def Stop(self):
    print("cat stop")

class Duck(Base):
  def Start(self):
    print("duck start")
  def Stop(self):
    print("duck stop")

class Puppy(Base):
  def Start(self):
    print("puppy start")
  def Stop(self):
    print("puppy stop")

def MY_Call(m):
  m.Start()
  m.Stop()

if __name__ == '__main__':
  MY_Call(Puppy())
  MY_Call(Cat())
  MY_Call(Duck())
```
