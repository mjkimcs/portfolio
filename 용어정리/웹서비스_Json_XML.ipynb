## 웹서비스
- 네트워크 상에서 서로 다른 종류의 컴퓨터들 간에 상호작용을 하기위한 서비스 지향적 분산 컴퓨팅 기술
- 데이터 교환 방식(객체 단위로 교환)  
      1. XML(eXtensible Markup Language) : 문서형식
      2. Json : dict/list/내부정보
- SOAP, WSDL, UDDI 등의 주요 표준 기술로 이루어진 소프트웨어 시스템
- 웹서비스 구성요소
      1. SOAP(Simple Object Access Protocol) : XML과 Http 등의 프로토콜을 기본으로 하여 다른 컴퓨터에 있는 데이터나 서비스를 호출하기 위한 통신규약
      2. WSDL(Web Service Description Language) : XML기반으로 작성, 사용법을 알려주는 인터페이스
      3. UDDI(Universal Description, Discovery and Integration) : 개방형표준과 비독점적인기술을 기반으로 개발된 전역 비즈니스 레지스터리로 다양한 웹서비스를 사용자가 검색해서 사용할 수 있는 요소

## Json 핵심정리
- dump() load() : 파일
- dumps() : list or dict -> json
- type: str
- 출력하면 더블쿼터("")

```python
s = "Hello, World!"

print(str(s), type(str(s)))
print(repr(s),type(repr(s)))
```

```python
import json

def Test01():
  data = [{"a":"A", "b":(2,4), "c":3.0}]
  print("data :", data, type(data))
  
  print("repr data :", repr(data), type(repr(data)))
  print("str :", str(data), type(str(data)))

  data_string = json.dumps(data, indent=4, sort_keys=True)
  print("json으로 변환 : ", data_string, type(data_string))


if __name__ == '__main__':
  Test01()
```

```python
# list
def Test02():
  data = [{"a":"A", "b":(2,4), "c":3.0}]

  encoded = json.dumps(data) # list -> json
  print("Encoded :", encoded, type(encoded))
  print("Repr :", repr(encoded), type(repr(encoded))) # json전체를 문자열로 변환

  decoded = json.loads(encoded) # json -> list
  print("Decoded :", decoded, type(decoded))


if __name__ == '__main__':
  Test02()
```

```python
# dict
def Test02():
  data = {"a":"A", "b":(2,4), "c":3.0}

  data_string = json.dumps(data) # dict -> json
  print("Encoded :", data_string, type(data_string))

  decoded = json.loads(data_string) # json -> dict
  print("Decoded :", decoded, type(decoded))


if __name__ == '__main__':
  Test02()
```

```python
# 문자열
def Test02():
  data = '{"a":"A", "b":(2,4), "c":3.0}' # 문자열로 인식
  print(data, type(data)) # 싱글쿼터('')가 없으면 dict

  data_string = json.dumps(data) # 문자열 -> json
  print("Encoded :", data_string, type(data_string))

  decoded = json.loads(data_string) # json -> 문자열
  print("Decoded :", decoded, type(decoded))


if __name__ == '__main__':
  Test02()
```

```python
# 문자열은 바로 json.loads() 가능
# 단, 값이 튜플형이면 불가능

# 파일이 아닐 때 + 클래스 문제
import json

My_s = '{"name": "민정", "sisters": ["민주", "태양"], "goal":"Happy"}'

class my_jsonObject:
  def __init__(self, d):
    self.__dict__ = d
    print(self.__dict__)
    print(d)


if __name__ == '__main__':
  data = json.loads(My_s, object_hook=my_jsonObject) # d에 My_s가 대입됨
  print("1.", data, type(data))
  print("2.", data.name)
  print("3.", data.sisters)
  for i in data.sisters:
    print("for loop", end=" ")
    print(i)
```

```python
# 파일 + 클래스 문제
import json

class Student:
  def __init__(self, d):
    self.__dict__ = d

def Prn02():
  data = json.load(fp=open("STUDENT.json", "r"), object_hook=Student)
  for i in data.STUDENT:
    hap = i.SCORE.KOR + i.SCORE.ENG + i.SCORE.MATH
    print(i.NAME , hap)


if __name__ == '__main__':
  Prn02()
```
